```python
# import modules needed
import pandas as pd
```

Clean data from the core metrics file

```python
# read the historical core metrics csv
c_df = pd.read_csv("../CSV Files/RDC_Inventory_Core_Metrics_Zip_History.csv")
c_df.head()
```

```python
# row count core metrics
c_row_count = c_df.shape[0]
c_row_count
```

```python
# data types for core metrics
c_df.info()
```

```python
# find the row values with non matching value types
non_date_values = c_df[~c_df['month_date_yyyymm'].astype(str).str.match(r'^\d{6}$')]
print(non_date_values['month_date_yyyymm'].unique())
```

```python
# Filter out the problematic row
c_df = c_df[c_df['month_date_yyyymm'] != 'quality_flag = 1:  year-over-year figures may be impacted']

# Convert "month_date_yyyymm" to a datetime format
c_df['month_date_yyyymm'] = pd.to_datetime(c_df['month_date_yyyymm'], format='%Y%m')

# Reformat the datetime column to the desired string format
c_df['month_date_yyyymm'] = c_df['month_date_yyyymm'].dt.strftime('%Y-%m')

c_df.head()
```

```python
# split the zip_name column into city and state columns
c_df[['city_name', 'state']] = c_df['zip_name'].str.split(', ', expand=True)
c_df = c_df.drop(columns=['zip_name'])
c_df.head()
```

```python
# Get the current column order
cols = c_df.columns.tolist()

# Reorder columns to place "state" and then "city_name" after "postal_code"
cols.remove('state')
cols.remove('city_name')
new_order = cols[:2] + ['state', 'city_name'] + cols[2:]

c_df = c_df[new_order]
```

```python
# convert the state column values to upper case
c_df['state'] = c_df['state'].str.upper()
```

```python
c_df = c_df.dropna(subset=['state'])
```

```python
c_df.info()
```

Clean data from the hotness metrics file

```python
# read the hotness metrics csv
h_df = pd.read_csv("../CSV Files/RDC_Inventory_Hotness_Metrics_Zip_History.csv")
h_df.head()
```

```python
# row count for hotness metrics
h_row_count = h_df.shape[0]
h_row_count
```

```python
# data types for hotness metrics
h_df.info()
```

```python
# find the row values with non matching value types
non_date_values = h_df[~h_df['month_date_yyyymm'].astype(str).str.match(r'^\d{6}$')]
print(non_date_values['month_date_yyyymm'].unique())
```

```python
# Filter out the problematic row
h_df = h_df[h_df['month_date_yyyymm'] != 'quality_flag = 1:  year-over-year figures may be impacted']

# Convert "month_date_yyyymm" to a datetime format
h_df['month_date_yyyymm'] = pd.to_datetime(h_df['month_date_yyyymm'], format='%Y%m')

# Reformat the datetime column to the desired string format
h_df['month_date_yyyymm'] = h_df['month_date_yyyymm'].dt.strftime('%Y-%m')

h_df.head()
```

```python
# split the zip_name column into city and state columns
h_df[['city_name', 'state']] = h_df['zip_name'].str.split(', ', expand=True)
h_df = h_df.drop(columns=['zip_name'])
h_df.head()
```

```python
# Get the current column order
cols = h_df.columns.tolist()

# Reorder columns to place "state" and then "city_name" after "postal_code"
cols.remove('state')
cols.remove('city_name')
new_order = cols[:2] + ['state', 'city_name'] + cols[2:]

h_df = h_df[new_order]

# convert the state column values to upper case
h_df['state'] = h_df['state'].str.upper()
```

```python
h_df = h_df.dropna(subset=['state'])
```

```python
h_df.info()
```

```python
# read the hotness metrics csv
zips_df = pd.read_csv("../CSV Files/uszips.csv")
zips_df.head()
```

```python
zips_df.info()
```

```python
# convert the zip data from the dataframes into integers
c_df['postal_code'] = c_df['postal_code'].astype(int)
h_df['postal_code'] = h_df['postal_code'].astype(int)
```

```python
# update column names to match the merge for zip & county names
zips_df.rename(columns={'county_fips':'postal_code'}, inplace=True)
```

```python
# map the zip codes to the county names
county_map = zips_df.set_index('postal_code')['county_name'].to_dict()

c_df['county_name'] = c_df['postal_code'].map(county_map)
h_df['county_name'] = h_df['postal_code'].map(county_map)
```

```python
# replace not a number entries with blanks
c_df['county_name'].fillna('', inplace=True)
h_df['county_name'].fillna('', inplace=True)
```

```python

# Save the finalized dataframes to new CSV files
c_df.to_csv('Core_Metrics_Zip_History.csv', index=False)
h_df.to_csv('Hotness_Metrics_Zip_History.csv', index=False)
```
