```python
import pandas as pd
```

```python
c_df = pd.read_csv("../CSV Files/Core_Metrics_Zip_History.csv")
c_df.head()
```

```python
h_df = pd.read_csv("../CSV Files/Hotness_Metrics_Zip_History.csv")
h_df.head()
```

```python
# merge the dataframes
combo_df = pd.merge(h_df, c_df, on=['postal_code', 'month_date_yyyymm'])
```

```python
combo_df
```

```python
combo_df.info()
```

```python
print(combo_df.columns)
```

```python
# delete duplicate columns shared by both dataframes
col_to_delete = ['county_name_x',
       'state_y', 'city_name_y', 'median_listing_price_y',
       'median_listing_price_mm_y', 'median_listing_price_yy_y', 'median_days_on_market_y',
       'median_days_on_market_mm_y', 'median_days_on_market_yy_y', 'quality_flag_y',
       'county_name_y']
combo_df = combo_df.drop(columns=col_to_delete)
combo_df.info()
```

```python
combo_df = combo_df.fillna(0)
combo_df.info()
```

```python
print(combo_df.columns)
```

```python
# create a new dataframe for only the ML necessary columns
# remove unnecessary columns
column_to_delete = ['quality_flag_x', 'median_listing_price_vs_us', 'median_dom_vs_us',
    'ldp_unique_viewers_per_property_mm', 'ldp_unique_viewers_per_property_yy', 
    'ldp_unique_viewers_per_property_vs_us', 'median_square_feet', 
    'median_square_feet_mm', 'median_square_feet_yy', 'state_x', 'city_name_x']

# delete colujns fro mthe dataframe
combo_df.drop(columns=column_to_delete, inplace=True)

combo_df.head()
```

```python
# convert the month_date_yyyymm column to a date formate then convert to a reference point of months to the stating date
combo_df['month_date_yyyymm'] = pd.to_datetime(combo_df['month_date_yyyymm'],format='%Y-%m')

# determine the earliest date from the dataset
early_date = combo_df['month_date_yyyymm'].min()

# calculate the # of months since the earliest date
combo_df['months_since_reference'] = (
    (combo_df['month_date_yyyymm'].dt.year - early_date.year) * 12 +
    combo_df['month_date_yyyymm'].dt.month - early_date.month
)

combo_df.head(20)
```
